# Module 1.1. Measures of Central Tendency

**Ukuran Pemusatan** adalah statistik deskriptif yang digunakan untuk menggambarkan pusat distribusi data. Ukuran pemusatan memberikan titik tengah atau "pusat" dari suatu distribusi data. Ada beberapa ukuran pemusatan yang umum digunakan, yaitu:

1. Mean (Rata-rata)
2. Median
3. Mode

## Case Study

Didalam sebuah ruangan, terdapat 12 orang yang sedang berkumpul dengan umur sebagai berikut:

25 | 12 | 20 | 20 | 15 | 21 | 22 | 19 | 19 | 24 | 17 | 65

Coba analisis pusat distribusi data, untuk melihat apakah ada outlier pada data tersebut.

## a. Mean (Rata-Rata)

**Mean atau rata-rata** adalah ukuran pemusatan yang paling umum digunakan, yaitu jumlah dari semua nilai dalam dataset dibagi dengan jumlah total nilai.

```python
# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

# Jumlah nilai dalam dataset 
jumlah_nilai = sum(data)
print(f"Jumlah semua nilai: {jumlah_nilai}")

# Jumlah total data dalam dataset 
jumlah_total_data = len(data)
print(f"Jumlah total data: {jumlah_total_data}")

# Jumlah semua nilai dalam dataset 
mean_value = jumlah_nilai / jumlah_total_data
print(f"Mean: {mean_value}")
```

Jika divisualisasikan dalam bentuk grafik, maka Mean akan terlihat sebagai berikut:

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Add a vertical line at the mean
plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Namun, terkadang Mean kurang mewakili pusat data yang sebenarnya. Terlebih jika ada data **"Outlier"**, yaitu nilai yang secara signifikan berbeda dari sebagian besar data dalam kumpulan sampel. Outlier dapat menjadi nilai yang sangat tinggi atau sangat rendah dibandingkan dengan nilai lain dalam kumpulan data.

Umur 60 tahun adalah outlier dalam sampel studi kasus. Coba perhatikan pada histogram di atas, terlihat jelas bahwa sebagian besar umur orang-orang di dalam ruangan terkonsentrasi sekitar 10 hingga 30 tahun. Namun, ada satu orang sepuh yang memiliki umur 60 tahun. 

Kehadiran outlier dapat mempengaruhi analisis statistik dan mengubah interpretasi dari distribusi data. Dalam beberapa kasus, outlier mungkin mengindikasikan kesalahan pengukuran atau kejadian yang tidak biasa. Oleh karena itu, penting untuk mengidentifikasi dan memahami outlier saat menganalisis data.

Salah satu masalah adanya outlier adalah dapat **mengacaukan Mean**. Garis putus-putus merah pada histogram mewakili umur rata-rata di ruangan tersebut, yaitu 23,25, lebih tinggi dari umur 9 dari 12 orang dalam ruangan. Outlier mendorong Mean terlalu ke kanan.

Dalam kasus seperti ini, median mungkin memberi kita analisis pusat distribusi data yang lebih baik.

## b. Median
**Median** adalah nilai tengah dalam dataset ketika data diurutkan dari terkecil hingga terbesar. Jika ada jumlah data genap, median adalah rata-rata dari dua nilai tengah.

```python
# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

# Mengurutkan data dari terkecil hingga terbesar
sorted_data = sorted(data)
print(f"Data berurut: {sorted_data}")

# Jumlah total data dalam dataset 
n = len(sorted_data)
print(f"Jumlah total data: {n}")

index_tengah = n//2 # double slash // is used for floor division

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah 
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1 = sorted_data[index_tengah]
    median2 = sorted_data[index_tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    median_value = (median1 + median2) / 2
    
# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    median_value = sorted_data[index_tengah]
    print(f"Nilai tengah: {median_value}")

print(f"Median: {median_value}")
```

Jika divisualisasikan dalam bentuk grafik, maka Median akan terlihat sebagai berikut:

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Add a vertical line at the mean
plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Perhatikan bahwa, dalam studi kasus ini, median menggambarkan umur rata-rata lebih akurat dibandingkan Mean.

## c. Mode (Modus)
**Modus** adalah nilai yang paling sering muncul dalam dataset. Perlu diingat bahwa sebuah dataset dapat memiliki banyak modus.

```python
# Import library Counter untuk menghitung modus
from collections import Counter

# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter) # ubah ke tipe data dictionary
max_counter = max(list(counter.values())) # mencari data yang memiliki counter paling banyak
print(f"Max Counter: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"Mode: {mode_value}")
```

Jika divisualisasikan dalam bentuk grafik, maka Modus akan terlihat sebagai berikut:

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Add a vertical line at the mean
plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {mode_value[1]:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Melihat dari histogram di atas, sekilas kita tidak bisa melihat perbedaan antara Modus dengan Median karena menghasilkan pusat data yang mirip. 

Jadi kapan kita harus menggunakan modus untuk menggambarkan pusat distribusi data?
